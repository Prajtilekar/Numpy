### use dtype

```python
np.array([10,20,30])
```

```python
#int to float: upcasting
np.array([10,20,30],dtype='float')
```

```python
np.array([10,20,30],dtype='complex')
```

```python
#if we have float values
np.array([2.3,4.66,7.80,9.10])
```

```python
#convert float to int: downcasting
np.array([2.3,4.66,7.80,9.10],dtype='int')
```

```python
#numeric: order of datatypes
# int--> float-->complex
```

```python
#implitcite typecasting
np.array([1,2,3,4,5.])
```

```python
np.array([1,2,35,5+6j])
```

```python
np.array([1,2,35,5+6j])
```

## np.zeros(): creates an array of 0's

```python
#1D
np.zeros(9) #9: total number of zeros
```

```python
#2D
np.zeros((2,4))#(2 rows,4 columns)
```

```python
#3D
np.zeros((3,2,2))#(3 blocks,2 rows,2 columns)
```

```python
np.zeros((3,2,2),dtype='int')
```

## np.full(): creates an array with constant value


```python
np.full(4,fill_value=99)
```

```python
#2D
np.full((4,4),fill_value=77)
```

```python
np.full((3,3),fill_value=22)
```
