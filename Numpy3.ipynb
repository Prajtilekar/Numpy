#np.linspace(): used to create a sequence of numbers with constant spacing or linear spacing between 2 numbers



```python
"""
Signature:
np.linspace(
    start,
    stop,
    num=50,
    endpoint=True,
    retstep=False,
    dtype=None,
    axis=0,
)
Docstring:
Return evenly spaced numbers over a specified interval.

Returns `num` evenly spaced samples, calculated over the
interval [`start`, `stop`].
"""
```

```python
np.linspace(1,15)
# 1 to 15 numbers, please note here 15 is included due to endpoint is TRUE i.e stop is inclusive, if exclusive then 15 will not be present.
# Also the difference between the two consecutive nos is same for all.
#def linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
```

```python
#if we want only 20 numbers betwen 1-15
np.linspace(1,10,num=20)
```

```python
#if we set endpoint=False
#then stop wont be a part of output, 15 wil not be present
np.linspace(1,15,num=20,endpoint=False)
```

```python
#to check difference btwn 2 values set retstep=True
np.linspace(1,15,num=20,retstep=True,endpoint=False)
```

#np.where: condition based conversion we can carry out

```python
p = np.arange(24).reshape(6,4)
p
```

```python
"""
Docstring:
where(condition, [x, y])

Return elements chosen from `x` or `y` depending on `condition`.

if condition is True return x else return y
"""
# return 0 if value<12
np.where(p<12,0,p)
```

```python
#value>=18 +ve and remaining -ve
np.where(p>=18,-p,p)
```

```python
# convert in 0 and 1 form
#value>10 ==> 0
# else 1
np.where(p>10,0,1)
```
